```python
import numpy as np
import matplotlib.pyplot as plt
import bioframe
from scipy import signal
```

```python
ref_genome=bioframe.fetch_chromsizes ('mm9')
```

```python
bins=bioframe.binnify(ref_genome,1000000)
bins
```

```python
print(bins.iloc[365]) #365th bin location
```

# SINGLE CELL (1 MB Res)

```python
contact_matrix=np.loadtxt('../GSE154353_1mb_single_cell_matrices_all/1mb_matrices_cluster/DPM6bot1.Odd2Bo1.Even2Bo61.10000.unweighted')
```

```python
contact_matrix.shape
```

```python
#Normalize
max_value=np.max(contact_matrix)
contact_matrix=contact_matrix/max_value
```

```python
plt.figure(figsize=(10, 10))
plt.imshow(contact_matrix, cmap='hot', interpolation='nearest')
plt.colorbar(extend='both')
plt.clim(0, 0.4);
plt.show()
```

```python
s1=1000
s2=s1
e1=1500
e2=e1
plt.figure(figsize=(10, 10))
plt.imshow(contact_matrix[s1:e1,s2:e2], cmap='hot', interpolation='nearest')
plt.colorbar(extend='both')
plt.clim(0, 0.4);
plt.xlabel(bins.iloc[s1]['chrom']+" "+str(bins.iloc[s1]['start'])+" - "+bins.iloc[e1]['chrom']+" "+str(bins.iloc[e1]['end']))
plt.ylabel(bins.iloc[s2]['chrom']+" "+str(bins.iloc[s2]['start'])+" - "+bins.iloc[e2]['chrom']+" "+str(bins.iloc[e2]['end']))
plt.show()
```

```python
plt.figure(figsize=(10, 10))
plt.imshow(contact_matrix[0:199,0:199], cmap='hot', interpolation='nearest')
plt.colorbar(extend='both')
plt.clim(0, 0.4);
plt.xlabel("Chromosome 1")
plt.ylabel("Chromosome 1")
plt.show()
```
